```python
def generate_dataset(files_list, n_fft=1024, hop_length=512, n_mels=64, frames=5, verbose = True):
    dims = n_mels * frames

    iteration = range(len(files_list))
    if verbose:
        iteration = tqdm(iteration)
        
    for index in iteration:
        signal, sr = load_sound_file(files_list[index])
        features = extract_signal_features(
            signal,
            sr,
            n_fft=n_fft,
            hop_length=hop_length,
            n_mels=n_mels,
            frames=frames,
        )

        if index == 0:
            dataset = np.zeros((features.shape[0] * len(files_list), dims), np.float32)

        dataset[
            features.shape[0] * index : features.shape[0] * (index + 1), :
        ] = features

    return dataset

def load_sound_file(path, mono=False, channel=0):
    signal, sr = librosa.load(path, sr=None, mono=mono)
    if signal.ndim < 2:
        sound_file = signal, sr
    else:
        sound_file = signal[channel, :], sr

    return sound_file

def plot_loss_per_epoch(history, model_name=None, file_name=None):
    if model_name is None:
        model_name = ""
    else:
        model_name += ": "
    
    plot_data = pd.DataFrame(history.history)
    plot_data['epoch'] = range(plot_data.shape[0])
    plot_data = plot_data.melt(value_vars= ['loss', 'val_loss'], id_vars='epoch', var_name='type')

    gg_train = (ggplot(plot_data, aes(x = 'epoch', y = 'value', color = 'type')) + 
        theme_bw() + 
        geom_line())
    
    return(gg_train)  

def build_files_list(root_dir, normal_dir="normal", abnormal_dir="abnormal"):
    normal_files = []
    abnormal_files = []

    for root, _, files in os.walk(top=os.path.join(root_dir)):
        for name in files:
            current_dir_type = root.split("/")[-1]
            if current_dir_type == normal_dir:
                normal_files.append(os.path.join(root, name))
            if current_dir_type == abnormal_dir:
                abnormal_files.append(os.path.join(root, name))

    return normal_files, abnormal_files

def extract_signal_features(
    signal, sr, n_fft=1024, hop_length=512, n_mels=64, frames=5
):
    mel_spectrogram = librosa.feature.melspectrogram(
        y=signal, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels
    )
    log_mel_spectrogram = librosa.power_to_db(mel_spectrogram, ref=np.max)
    features_vector_size = log_mel_spectrogram.shape[1] - frames + 1
    
    dims = frames * n_mels

    if features_vector_size < 1:
        return np.empty((0, dims), np.float32)

    features = np.zeros((features_vector_size, dims), np.float32)
    for time in range(frames):
        features[:, n_mels * time : n_mels * (time + 1)] = log_mel_spectrogram[
            :, time : time + features_vector_size
        ].T

    return features
```

# Model for anomaly detection

```python
class anomaly_detection_AE():
    """
    Framework for anomaly detection using different methods for normal state modeling
    """
    def __init__(self, save_model = False, model_path = None, image_path = None):
        
        self.save_model = save_model
        self.model_path = model_path
        self.image_path = image_path
    
    def AE(self, input_dims, model_name=None):
        input_layer = Input(shape=(input_dims,))
        output = Dense(4*64, activation="relu")(input_layer)
        output = Dense(3*64, activation="relu")(output)
        output = Dense(3*64, activation="relu")(output)
        output = Dense(10*8, activation="relu")(output)
        output = Dense(3*64, activation="relu")(output)
        output = Dense(3*64, activation="relu")(output)
        output = Dense(4*64, activation="relu")(output)
        output = Dense(input_dims, activation=None)(output)

        return Model(inputs=input_layer, outputs=output, name=model_name)
    
    def train(self, data, model_name, plot = False, batch_size = 1024 * 4):
        """
        initializes and trains the autoencoder with the training data
        the train_data is splitted in train and validation to controll the train process
        
        
        :param train_data: numpy array, where the first dimension are the samples
        """
        
        self.samples = np.shape(data)[0]
        self.dim_1 = np.shape(data)[1]
        self.dim_2 = np.shape(data)[2]
        
        ind_train, ind_val = train_test_split(range(self.samples), train_size=0.9)
        data_train = data[ind_train,:,:].reshape((len(ind_train)*self.dim_1, self.dim_2))
        data_val = data[ind_val,:,:].reshape((len(ind_val)*self.dim_1,self.dim_2))   
        
        #self.mean = np.mean(data_train)
        #self.std = np.std(data_train)

        self.mean = data_train.mean(axis=0)
        self.std = data_train.std(axis=0)
        
        data_train = (data_train - self.mean) / self.std
        data_val = (data_val - self.mean) / self.std
        
        epochs = 200
        
        self.model = self.AE(self.dim_2, model_name)
        if plot:
            print(self.model.summary())
        
        self.model.compile(optimizer=Adam(learning_rate=1e-04),
                           loss="mean_squared_error")
        
        history = self.model.fit(data_train, data_train,
                                 batch_size=batch_size, epochs=epochs, verbose=plot,
                                 callbacks=[EarlyStopping(monitor="val_loss", min_delta = 0.0001, patience=5 )],
                                 validation_data=(data_val, data_val),
                                 shuffle=True
                                )
        
        if self.save_model:
            self.model.save(os.path.join(self.model_path, model_name + ".h5"))
        
        if plot:
            gg_loss = plot_loss_per_epoch(history, model_name=model_name)
            gg_loss.draw()
        
        return
    
    def predict(self, data):
        data = (data - self.mean) / self.std
        pred = self.model.predict(data, batch_size = data.shape[0], verbose=False)
        rec_diff = np.square(pred - data)
        rec_diff_reshape = rec_diff.reshape(int(pred.shape[0]/self.dim_1),self.dim_1,self.dim_2)
        recon_errors = rec_diff_reshape.mean(axis=(1,2))
        return recon_errors
    
class anomaly_detection_SV():
    """
    Framework for anomaly detection using different methods for normal state modeling
    """
    def __init__(self, save_model = False, model_path = None, image_path = None, ratio = 0.25):
        self.save_model = save_model
        self.model_path = model_path
        self.ratio = ratio
    
    def train(self, data, model_name, plot = False, batch_size = 1024 * 3):
        """
        :param train_data: numpy array, where the first dimension are the samples
        """
        
        self.samples = np.shape(data)[0]
        self.dim_1 = np.shape(data)[1]
        self.dim_2 = np.shape(data)[2]
        
        data_train = data.reshape((self.samples*self.dim_1, self.dim_2)) 
        
        self.mean = data_train.mean(axis=0)
        self.std = data_train.std(axis=0)
        
        data_train = (data_train - self.mean) / self.std

        self.model = linear_model.SGDOneClassSVM(nu = self.ratio, tol=1e-4)
        
        self.model.fit(data_train)
        
        if self.save_model:
            pickle.dump(self.model, open(os.path.join(self.model_path, model_name + ".h5"), 'wb'))
                                                                                 
        return
    
    def predict(self, data):
        data = (data - self.mean) / self.std

        score = self.model.score_samples(data)
        score = score.reshape(int(score.shape[0]/self.dim_1),self.dim_1)
        recon_errors = score.mean(axis=1) ** 2
        return recon_errors
    
class anomaly_detection_PCA():
    """
    Framework for anomaly detection using different methods for normal state modeling
    """
    def __init__(self, save_model = False, model_path = None, image_path = None):
        self.save_model = save_model
        self.model_path = model_path
    
    def train(self, data, model_name, plot = False, batch_size = 1024 * 3):
        """
        :param train_data: numpy array, where the first dimension are the samples
        """
        
        self.samples = np.shape(data)[0]
        self.dim_1 = np.shape(data)[1]
        self.dim_2 = np.shape(data)[2]
        
        data_train = data.reshape((self.samples*self.dim_1, self.dim_2)) 
        
        self.mean = data_train.mean(axis=0)
        self.std = data_train.std(axis=0)
        
        data_train = (data_train - self.mean) / self.std
        
        self.model = PCA(5)
        
        self.model.fit(data_train)
        print(np.sum(self.model.explained_variance_ratio_))
        
        if self.save_model:
            pickle.dump(self.model, open(os.path.join(self.model_path, model_name + ".h5"), 'wb'))
                                                                                 
        return
    
    def predict(self, data):
        data = (data - self.mean) / self.std
        pred = self.model.inverse_transform(self.model.transform(data))
        rec_diff = np.square(pred - data)
        rec_diff_reshape = rec_diff.reshape(int(pred.shape[0]/self.dim_1),self.dim_1,self.dim_2)
        recon_errors = rec_diff_reshape.mean(axis=(1,2))
        
        return recon_errors
    
 
```

# Data Splitting and Preprocessing

## Data loading and feature generation

```python
# Load and process all files
# feature parameters
n_fft = 2048
hop_length = 512
n_mels = 64
frames = 1

dim_1 = 313 - frames + 1
dim_2 = n_mels * frames


normal_files, abnormal_files = build_files_list(root_dir=file_path)
# only load part of the data JZ
if partialload_Flag == True:
    normal_files = normal_files[0:int(len(abnormal_files) * 1.2)]
    #print(len(normal_files))
    #abnormal_files = abnormal_files[::3]

normal_labels = np.zeros(len(normal_files))
abnormal_labels = np.ones(len(abnormal_files))

normal_data = generate_dataset(
        normal_files, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, frames=frames, verbose = tqdm_installed)
abnormal_data = generate_dataset(
        abnormal_files, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, frames=frames, verbose = tqdm_installed)
```

## reshape and shuffle samples

```python
## reshape
normal_data_resh = normal_data.reshape(int(normal_data.shape[0]/dim_1),dim_1,dim_2)
abnormal_data_resh = abnormal_data.reshape(int(abnormal_data.shape[0]/dim_1),dim_1,dim_2)

data_resh = np.concatenate([normal_data_resh,abnormal_data_resh], axis =0)
data_label = np.concatenate([normal_labels,abnormal_labels], axis =0)

data = data_resh.reshape((data_resh.shape[0]*dim_1, dim_2))

#data frame to collect all results
results = pd.DataFrame({'label' : data_label})

# Create splits
normal_ind = np.arange(len(normal_files))
abnormal_ind = np.arange(len(abnormal_files)) + normal_ind.shape[0]

# Index all samples
all_ind = np.hstack([normal_ind,abnormal_ind])
results['ind'] = all_ind

# Number of samples
N = all_ind.shape[0]

# shuffle to shuffle file strucure
normal_ind_shuffle = normal_ind.copy()
abnormal_ind_shuffle = abnormal_ind.copy()

if shuffle:
    np.random.shuffle(normal_ind_shuffle)
    np.random.shuffle(abnormal_ind_shuffle)

# insert abnormal in middle
cut_ind = int(np.floor(normal_ind.shape[0]/2))
normal_ind_a = normal_ind_shuffle[:cut_ind]
normal_ind_b = normal_ind_shuffle[cut_ind:]
abnormal_toadd = abnormal_ind_shuffle.copy()

# insert 
all_ind_shuffle = np.hstack([normal_ind_a,abnormal_toadd,normal_ind_b])


results['ind_shuffle'] = all_ind_shuffle
ratio = len(abnormal_ind) / len(all_ind)
```
